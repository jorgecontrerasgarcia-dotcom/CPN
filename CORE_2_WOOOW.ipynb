```python
#EJERCICIO 2 DE CORE 2 SIII (estoy esuchando radiohead y ando en media filosofia)
cantantes = [
    {"nombre": "Thom Yorke", "pais": "Inglaterra"},
    {"nombre": "Colin Greenwood", "pais": "inglaterra"},
    {"nombre": "Ed O'Brien", "pais": "Inglaterra"},
    {"nombre": "Jonny Greenwood", "pais": "Oxford"}
]

def iterarDiccionario(lista):

    for diccionario in lista:

        pares = []
        for llave, valor in diccionario.items():
            pares.append(f"{llave} - {valor}")

        print(", ".join(pares))

iterarDiccionario(cantantes)
```
